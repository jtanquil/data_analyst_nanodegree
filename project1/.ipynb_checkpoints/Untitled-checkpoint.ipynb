```python
import pandas as pd
```

```python
df = pd.read_csv("results.csv")
```

```python
df.head()
```

```python
df = df.sort_values(by = ["city", "year"])
```

```python
df.head()
```

```python
test = pd.concat([df.loc[df.city == city, ["avg_temp"]].rolling(5).mean() for city in df.city.unique()]).rename(columns = {"avg_temp" : "5yr_rolling"})
```

```python
test
```

```python
df = pd.merge(df, test, left_index = True, right_index = True)
```

```python
%matplotlib inline
```

```python
pivot = df.pivot(index = "year", columns = "city", values = "5yr_rolling")
```

```python
pivot.plot(figsize = (20, 10))
```

```python
def get_moving_avg(s, period):
    return s.rolling(period).mean()
```

```python
global_test = pd.merge(global_temp, get_moving_avg(global_temp.avg_temp, 5), left_index = True, right_index = True).rename(columns = {"avg_temp_y" : "5yr_moving_avg"})
```
