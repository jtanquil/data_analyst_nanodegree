```python
# get unique genres
genres = []
```

```python
def parse_genres(s: str, sep: str, genre_list: list):
    separated = str(s).split(sep)
    genre_list += separated
```

```python
df["genres"].apply(lambda x: parse_genres(x, "|", genres))
```

```python
unique_genres = set(genres)
```
