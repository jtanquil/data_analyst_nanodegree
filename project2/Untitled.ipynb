```python
import matplotlib.pyplot as plt
import seaborn as sns
import csv
```

```python
df = pd.read_csv("cleaned_data.csv")
```

```python
def csv_to_list(name: str):
    with open(name) as f:
        rd = csv.reader(f)
        l = []
        
        for row in rd:
            l += row
            
        return l
```

```python
genres = csv_to_list("genres.csv")
```

```python
studios = csv_to_list("studios.csv")
```

```python
cast = csv_to_list("cast.csv")
```

```python
leads = csv_to_list("leads.csv")
```

```python
genre_revs = {}
```

```python
# gets adjusted revenue for movies in each genre
for genre in genres:
    genre_revs[genre] = df.revenue_adj[df.genres.apply(lambda x: genre in str(x))]
```

```python
for genre in genre_revs.keys():
    print(genre + ": $" + '%.2f' % genre_revs[genre].mean())
```

```python
genre_means = {genre:genre_revs[genre].mean() for genre in genre_revs.keys()}
```

```python
genre_mean_df = pd.DataFrame({"means": list(genre_means.values())}, index = genre_means.keys())
```

```python
from matplotlib.ticker import StrMethodFormatter
```

```python
ax = genre_mean_df.sort_values(by = ["means"]).plot(kind = "bar", figsize = (15, 10))
ax.yaxis.set_major_formatter(StrMethodFormatter('${x:,.0f}'))
ax.set_title("Average Movie Revenue by Genre")
ax.set_ylabel("Adjusted Revenue")
ax.set_xlabel("Genre")
# add average adjusted revenue for all movies in the df
l = ax.axhline(y = df.revenue_adj.mean())
l.set_linestyle("--")
l.set_color("red")
# change the legend
ax.legend(handles = [l], labels = ["Average For All Genres"], loc = 2)
```

```python
df.groupby(["release_year"])["revenue_adj"].mean().rolling(10).mean().plot()
```

```python
# check number of films released each year
df.groupby(["release_year"])["revenue_adj"].count().plot()
```

```python
df["release_date"] = pd.to_datetime(df["release_date"])
```

```python
df["month"] = df["release_date"].apply(lambda x: x.month)
```

```python
ax = df.groupby(["month"])["revenue_adj"].mean().plot(figsize = (15, 10))
ax.yaxis.set_major_formatter(StrMethodFormatter('${x:,.0f}'))
ax.set_xticks(range(1, 13))
ax.set_xticklabels(["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"])
ax.set_title("Average Movie Revenue per Month")
ax.set_ylabel("Adjusted Revenue")
ax.set_xlabel("Month")
# add average adjusted revenue for all movies in the df
l = ax.axhline(y = df.revenue_adj.mean())
l.set_linestyle("--")
l.set_color("red")
# change the legend
ax.legend(handles = [l], labels = ["Average For All Months"], loc = 2)
```
